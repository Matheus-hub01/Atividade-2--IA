# Atividade 2 - AULA 02 - IA


Você recebeu uma lista contendo as idades de pacientes atendidos em uma clínica.


Lista utilizada no exercício:



idades = [23, 45, 31, 54, 29, 61, 40, 37, 48, 33]



1) Mostre:

a) A primeira idade

b) A última idade

c) O tamanho da lista

d) A terceira idade



```python
idades = [23, 45, 31, 54, 29, 61, 40, 37, 48, 33]


print("a) Primeira idade:", idades[0])


print("b) Última idade:", idades[-1])


print("c) Tamanho da lista:", len(idades))


print("d) Terceira idade:", idades[2])
```

2) Calcule manualmente:

a) A média das idades

b) A maior idade

c) A menor idade



```python
idades = [23, 45, 31, 54, 29, 61, 40, 37, 48, 33]

soma = sum(idades)
media = soma / len(idades)
maior = max(idades)
menor = min(idades)

print("a) A Soma de todas as idades é =", soma,"e dividindo pelos valores (10) a Média fica: =", media, "anos.")

print("b) Maior idade =", maior, "anos")

print("c) Menor idade =", menor, "anos")
```

3) Crie novas listas contendo:

a) Idades acima de 40

b) Idades entre 30 e 50

c) Idades pares



```python
idades = [23, 45, 31, 54, 29, 61, 40, 37, 48, 33]


acima_40 = [idade for idade in idades if idade > 40]


entre_30_50 = [idade for idade in idades if 30 <= idade <= 50]


pares = [idade for idade in idades if idade % 2 == 0]

print("a) Idades acima de 40:", acima_40)
print("b) Idades entre 30 e 50:", entre_30_50)
print("c) Idades pares:", pares)
```

4) Crie uma nova lista com todas as idades aumentadas em 5 anos.



```python
idades = [23, 45, 31, 54, 29, 61, 40, 37, 48, 33]

idades_mais_5 = [idade + 5 for idade in idades]

print("Nova lista com as idades + 5 anos :", idades_mais_5)
```

5) Crie uma lista chamada grupo_risco com idades acima de 50.
Mostre quantos pacientes estão no grupo de risco.

```python
idades = [23, 45, 31, 54, 29, 61, 40, 37, 48, 33]

grupo_risco = [idade for idade in idades if idade > 50]

print("Grupo de risco:", grupo_risco)
print("Quantidade de pacientes no grupo de risco:", len(grupo_risco))
```
